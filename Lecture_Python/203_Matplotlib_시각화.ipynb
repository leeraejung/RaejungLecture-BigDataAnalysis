# 1. Matplotlib

```python
# Visual Python: Data Analysis > Import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Visual Python: Data Analysis > File
df = pd.read_csv('./data/iris.csv')
df
```

### 1.1 Line chart

```python
plt.plot(df['sepal.length'])
plt.show()
```

### 1.2 Bar chart

```python
df_val = df['variety'].value_counts().reset_index()
df_val
```

```python
# Visual Python: Visualization > Matplotlib
plt.bar(df_val['index'], df_val['variety'])
plt.show()
```

### 1.3 Scatter chart

```python
# Visual Python: Visualization > Matplotlib
plt.scatter(df['sepal.length'], df['petal.length'])
plt.show()
```

# 2. Pandas plot

### 2.1 Line chart

```python
df.plot()
plt.show()
```

### 2.2 Bar chart

```python
# Visual Python: Library > Pandas Plot
df_val.plot(kind='bar')
plt.show()
```

# 3. Seaborn

### 3.1 Scatter plot

```python
# Visual Python: Visualization > Seaborn
sns.scatterplot(data=df)
plt.show()
```

### 3.2 Bar plot

```python
# Visual Python: Visualization > Seaborn
sns.barplot(data=df)
plt.show()
```

### 3.3 Hist plot

```python
# Visual Python: Visualization > Seaborn
sns.histplot(data=df, x='sepal.length')
plt.show()
```
